```python
import math
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.linspace(-math.pi, math.pi, 1000)
y = torch.sin(x)
```

```python
a = torch.randn(())
b = torch.randn(())
c = torch.randn(())
d = torch.randn(())
```

```python
y_random = a * x**3 + b* x**2 + c*x +d
plt.subplot(2,1,1)
plt.title("y True")
plt.plot(x,y)

plt.subplot(2,1,2)
plt.title("y prediction")
plt.plot(x,y_random)
plt.subplots_adjust(hspace=0.5)

plt.show()
```
