```python
!pip install datasets
from datasets import load_dataset

# 데이터셋 로드 () Naver 뉴스 요약 데이터셋)
dataset = load_dataset("daekeun-ml/naver-news-summarization-ko")
print(dataset)

# 데이터 전처리 함수 정의
def preprocess_data(example):
    model_inputs = tokenizer(
        example["document"],
        max_length=512,
        padding="max_length",
        truncation=True
    )
    labels = tokenizer(
        example["summary"],
        max_length=128,
        padding="max_length",
        truncation=True
    )
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# 데이터셋 전처리
tokenized_dataset = dataset.map(
    preprocess_data,
    batched=True,
    remove_columns=dataset["train"].column_names
)

# 결과 확인
print(tokenized_dataset)
```

```python
from torch.utils.data import DataLoader
from transformers import DataCollatorForSeq2Seq

seq2seq_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    padding="longest",
    return_tensors="pt"
)

# DataLoader 구성
seq2seq_dataloader = DataLoader(
    tokenized_dataset["train"],  # tokenized_dataset으로 수정
    collate_fn=seq2seq_collator,
    batch_size=4,
    shuffle=False
)

# 배치 확인
seq2seq_iterator = iter(seq2seq_dataloader)
seq2seq_batch = next(seq2seq_iterator)
for key, value in seq2seq_batch.items():
    print(f"{key} : {value.shape}")
```

```python
!pip install optuna

import optuna
from transformers import Seq2SeqTrainingArguments, Seq2SeqTrainer

# Optuna 최적화 함수 정의
def objective(trial):
    # 최적화할 하이퍼파라미터
    batch_size = trial.suggest_categorical("batch_size", [4, 8, 16])
    learning_rate = trial.suggest_float("learning_rate", 1e-5, 5e-4, log=True)
    num_train_epochs = trial.suggest_int("num_train_epochs", 1, 5)

    # 학습 설정
    training_args = Seq2SeqTrainingArguments(
        output_dir="./kobart-optimized",
        per_device_train_batch_size=batch_size,
        learning_rate=learning_rate,
        num_train_epochs=num_train_epochs,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        logging_dir="./logs",
        predict_with_generate=True,
        fp16=True,
        gradient_accumulation_steps=4
    )

    # Trainer 정의
    trainer = Seq2SeqTrainer(
        model=model,
        args=training_args,
        train_dataset=tokenized_dataset["train"],  # tokenized_dataset 활용
        eval_dataset=tokenized_dataset["validation"],  # tokenized_dataset 활용
        tokenizer=tokenizer
    )

    # 학습 및 평가
    trainer.train()
    metrics = trainer.evaluate()
    return metrics["eval_loss"]

# Optuna 실행
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=10)

# 최적의 하이퍼파라미터 출력
print("Best hyperparameters:", study.best_params)
```

```python
# 평가 데이터 준비
test_input = " 대한민국의 수도는 어디인가요?"
input_tokenized = tokenizer(
    test_input,
    max_length=512,
    return_tensors="pt",
    truncation=True,
    padding="max_length"
)

input_tokenized["input_ids"] = input_tokenized["input_ids"].to(model.device)


# 모델 예측
output = model.generate(input_tokenized["input_ids"], max_length=50, num_beams=4)
print("모델 출력:", tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
# 최적의 하이퍼파라미터 확인
print("Best hyperparameters:", study.best_params)

# 최적의 Validation Loss 확인
print("Best Validation Loss:", study.best_value)

# 전체 Trial 기록 확인
for trial in study.trials:
    print(f"Trial {trial.number}: Loss = {trial.value}, Params = {trial.params}")
```
